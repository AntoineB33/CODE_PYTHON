```python
##Fonctions utiles

#Create sub groups
# in : action partition
# out : 3 binary matrix, one for each frequency
def create_subgroups(partition):
    sub6 = np.zeros(len(partition))
    mmW = np.zeros(len(partition))
    unpartitioned = np.zeros(len(partition))
    
    for i in range (len(partition)):
        if(partition[i] == 1):
            sub6[i] = 1
        elif(partition[i] == 2):
            mmW[i] = 1
        else:
            unpartitioned[i] = 1
            
    #Transform in integer arrays
    sub6 = sub6.astype(int)
    mmW = mmW.astype(int)
    unpartitioned = unpartitioned.astype(int)
    
    return sub6, mmW, unpartitioned

#Sample SINR matrix for max_epoch frames
def datasampler(SINR, max_epoch, K): 
    max_size = len(SINR[0])
    mask = np.random.randint(0, max_size, max_epoch)
    #mask = np.arange(40, 50)
    
    res = []
    for i in range(int(K)):
        res.append(SINR[i][mask])
        
    return res
                    
#Create user infos (los, rate required, error req, delay req) 
def users_info_freq(partition, K):
    s6 = []
    mmW = []
    for i in range(K):
        los = users_infos[0][i]
        rate_req = users_infos[1][i]
        error_req = users_infos[6][i]
        delay_req = users_infos[5][i]
        if(partition[i] == 1):    
            s6.append([los, rate_req, error_req, delay_req])
        if(partition[i] == 2):
            mmW.append([los, rate_req, error_req, delay_req])
    
    return s6, mmW
```

```python
def F(schedule, SINR, K):
    size = len(schedule)
    nbBlock = int(size/K)
    capacity = np.zeros(size)
    for i in range (K):
        capacity[i*nbBlock:(i+1)*nbBlock] = schedule[i*nbBlock:(i+1)*nbBlock] * np.log(1+SINR)[i]
    return np.sum(capacity)

#Helper function to compute the Capacity (F value) of each user and stored in a size
#K_sub6 array
def F_k(schedule, SINR, K):
    size = len(schedule)
    nbBlock = int(size/K)
    capacity = np.zeros(size)
    res = []
    for i in range (K):
        capacity[i*nbBlock:(i+1)*nbBlock] = schedule[i*nbBlock:(i+1)*nbBlock] * np.log(1+SINR)[i]
        res.append(np.sum(capacity[i*nbBlock:(i+1)*nbBlock]))
    return np.array(res)

def V(schedule, SINR, Q_inv, l, K):
    size = len(schedule)
    nbBlock = int(size/K)
    
    dispersion = np.ones(size)
    square = np.zeros(K)
    for i in range(K):
        dispersion[i*nbBlock:(i+1)*nbBlock] = schedule[i*nbBlock:(i+1)*nbBlock] * np.log(e)**2 * (1 - 1/((1 + SINR)**2))[i]
        square[i] = np.sqrt(np.sum(dispersion[i*nbBlock:(i+1)*nbBlock]))
        
    dispersion_user_k = (Q_inv/np.sqrt(l)) * square 
    
    return np.sum(dispersion_user_k)

#Helper function to compute the Dispersion (V value) of each user and stored in a size
#K_sub6 array
def V_k(schedule, SINR, Q_inv, l, K):
    size = len(schedule)
    nbBlock = int(size/K)
    
    dispersion = np.ones(size)
    square = np.zeros(K)
    for i in range(K):
        dispersion[i*nbBlock:(i+1)*nbBlock] = schedule[i*nbBlock:(i+1)*nbBlock] * np.log(e)**2 * (1 - 1/((1 + SINR)**2))[i]
        square[i] = np.sqrt(np.sum(dispersion[i*nbBlock:(i+1)*nbBlock]))
        
    dispersion_user_k = (Q_inv/np.sqrt(l)) * square 
    
    return dispersion_user_k
```

```python
def W(schedule):
    return np.sum(schedule, axis = (0))

def E(schedule):
    schedule_squared = schedule * schedule
    return np.sum(schedule_squared, axis = (0))

def U(schedule, SINR, Q_inv, l, K):
    return - F(schedule, SINR, K) + V(schedule, SINR, Q_inv, l, K) + beta*(W(schedule) - E(schedule))
```

```python
def gradE(schedule, schedule_init):
    
    double_sched = 2*schedule
    difference = schedule - schedule_init
    whole = double_sched*difference
    return np.sum(whole, axis = (0))

def inv_ifNotZero(arr):
    return [1/x if x != 0 else 0 for x in arr]

def gradV(schedule, schedule_init, SINR, Q_inv, l, K):
    size = len(schedule)
    nbBlock = int(size/K)
    
    dispersion = np.ones(size)
    square = np.zeros(K)
    for i in range(K):
        dispersion[i*nbBlock:(i+1)*nbBlock] = schedule_init[i*nbBlock:(i+1)*nbBlock] * np.log(e)**2 * (1 - 1/((1 + SINR)**2))[i]
        square[i] = np.sqrt(np.sum(dispersion[i*nbBlock:(i+1)*nbBlock]))
    
    gradient_leftPart = (Q_inv/np.sqrt(l)) * inv_ifNotZero(square)
    
    gradient_rightPart = np.array( [(1/2) * np.log(e)**2 * (1 - 1/((1+ x)**2)) for x in SINR])
    
    gradMatrix_oneFrame = np.array( [gradient_leftPart[idx]*perUser_rightPart for idx,perUser_rightPart in enumerate(gradient_rightPart)])
    
    difference = schedule - schedule_init
    
    full_gradMatrix = np.zeros(size)
    for i in range(K):
        full_gradMatrix[i*nbBlock:(i+1)*nbBlock] = difference[i*nbBlock:(i+1)*nbBlock] * gradMatrix_oneFrame[i]
    
    return np.sum(full_gradMatrix)

#Helper function to compute the Dispersion (V value) of each user and stored in a size
#K_sub6 array
def gradV_k(schedule, schedule_init, SINR, Q_inv, l, K):
    size = len(schedule)
    nbBlock = int(size/K)
    
    dispersion = np.ones(size)
    square = np.zeros(K)
    for i in range(K):
        dispersion[i*nbBlock:(i+1)*nbBlock] = schedule_init[i*nbBlock:(i+1)*nbBlock] * np.log(e)**2 * (1 - 1/((1 + SINR)**2))[i]
        square[i] = np.sqrt(np.sum(dispersion[i*nbBlock:(i+1)*nbBlock]))
    
    gradient_leftPart = (Q_inv/np.sqrt(l)) * inv_ifNotZero(square)
    
    gradient_rightPart = np.array( [(1/2) * np.log(e)**2 * (1 - 1/((1+ x)**2)) for x in SINR])
    
    gradMatrix_oneFrame = np.array( [gradient_leftPart[idx]*perUser_rightPart for idx,perUser_rightPart in enumerate(gradient_rightPart)])
    
    difference = schedule - schedule_init
    
    full_gradMatrix = np.zeros(size)
    res = []
    for i in range(K):
        full_gradMatrix[i*nbBlock:(i+1)*nbBlock] = difference[i*nbBlock:(i+1)*nbBlock] * gradMatrix_oneFrame[i]
        res.append(np.sum(full_gradMatrix[i*nbBlock:(i+1)*nbBlock]))

    return np.array(res)
```

```python
def U_bar(schedule, schedule_init, SINR, Q_inv, l, K):
    return -F(schedule, SINR, K) + V(schedule_init, SINR, Q_inv, l, K) + gradV(schedule, schedule_init, SINR, Q_inv, l, K) + beta*(W(schedule) - E(schedule_init) - gradE(schedule, schedule_init))  
```

```python
## CONSTRAINTS : 
def C_delay(schedule, K, user_infos):
    cpt = 0
    res = []
    size = len(schedule)
    nbBlock = int(size / K)
    for k in range(K):
        for i in range(cpt, cpt+nbBlock, 5):
        
            for idx, x in enumerate(schedule[i:i+schedule_frameTime]):
                if(idx + 1 > user_infos[k,3]):
                    res.append(int(schedule[idx+i]))
        cpt += nbBlock
    return -np.sum(res)

def C_rate(schedule, schedule_init, SINR, Q_inv, l, K, user_infos):
    return F_k(schedule, SINR, K) - V_k(schedule_init, SINR, Q_inv, l, K) - gradV_k(schedule, schedule_init, SINR, Q_inv, l, K) - user_infos[:,1]
    
def C_relaxation(schedule):
    return E(schedule) - W(schedule)
    
def C_blockPerUser(schedule, N, K):
    nb_case = schedule_frameTime * N
    res = []
    temp = 0
    for i in range (nb_case):
        for k in range(K):
            temp += schedule[i + nb_case*k]
        res.append(1 - temp)         
        temp = 0
    return res 
    
    
def C_blockPerFrame(schedule):
    res = []
    size = len(schedule)
    for i in range (schedule_frameTime):
        res.append(2 - np.sum(schedule[i:size:schedule_frameTime]))
    return res

def C_zero(schedule):
    return schedule

def C_one(schedule):
    return 1-schedule 

def transform_array(schedule, K, N):
    array = np.zeros((K,N,5))
    cpt = 0
    for i in range(K):
        for j in range(N):
            for k in range(schedule_frameTime):
                array[i][j][k] = schedule[cpt]
                cpt += 1
    return array


def AllCons(schedule_init, SINR, Q_inv, l, K, N, users_infos):
    cons = []
    #Bounds : 
    cons += [{'type' : 'ineq', 'fun': C_zero}, {'type' : 'ineq', 'fun': C_one}]

    c_1 = {'type' : 'ineq', 'fun': C_relaxation}
    cons += [c_1]

    #Bloc User
    c_2 = {'type' : 'ineq', 'fun': C_blockPerUser, 'args':(N, K)}
    cons += [c_2]  

    #Block Frame
    c_3 = {'type' : 'ineq', 'fun' : C_blockPerFrame}
    cons += [c_3]

    #Delay
    c_delay = {'type':'ineq', 'fun' : C_delay, 'args':(K, users_infos)}
    cons += [c_delay]

    #Rate
    c_rate = {'type':'ineq', 'fun': C_rate, 'args': (schedule_init, SINR, Q_inv, l, K, users_infos)}
    cons += [c_rate]
    
    return cons
```

```python
def scheduler(partition):
    
    start_time = time.time()
    
    #Create sub groups (Sub-6, mmWave, unpartitioned)
    sub6_users, mmW_users, unpartitioned_users = create_subgroups(partition)
    
    K_total = len(partition)
    K_sub6 = sum(sub6_users)
    K_mmW = sum(mmW_users)
    
    #Manipulate SINR matrix to only keep s6 users (or mmW)
    SINR_sub6users_allframes = []
    SINR_mmWusers_allframes = []
    for i in range (K_total):
        if(partition[i] == 1):
            SINR_sub6users_allframes.append(SINR_S6_allframes[:,i])
        if(partition[i] == 2):
            SINR_mmWusers_allframes.append(SINR_mmW_allframes[:,i])
        
    #Sample MaxEpoch frames for each band
    if K_sub6 != 0:
        SINR_sub6 = datasampler(SINR_sub6users_allframes, max_epoch, K_sub6)
        SINR_sub6 = np.array(SINR_sub6)
    if K_mmW != 0:
        SINR_mmW = datasampler(SINR_mmWusers_allframes, max_epoch, K_mmW)
        SINR_mmW = np.array(SINR_mmW)
        

    #Create all the informations about the subgroup necessary for the scheduler
    users_infos_sub6, users_infos_mmW = users_info_freq(partition, K_total)
    Q_inv_sub6 = np.array([np.sqrt(2) * special.erfinv(1 - 2*x[2]) for x in users_infos_sub6])
    Q_inv_mmW = np.array([np.sqrt(2) * special.erfinv(1 - 2*x[2]) for x in users_infos_mmW])
    
    #Randomly put SINR at 0 for mmW users depending on LoS probability (if LoS = 1, then all SINR are kept to their value)
    for i in range(K_mmW):
        user_los_proba = users_infos_mmW[i][0]
        temp_rand_array = np.random.uniform(0, 1, max_epoch)
        for j in range (max_epoch):
            if(temp_rand_array[j] > user_los_proba):
                SINR_mmW[i][j] = 0. 

    users_infos_sub6 = np.array(users_infos_sub6)
    users_infos_mmW = np.array(users_infos_mmW)
                
    #Initialize the mean achieved rate & the probability arrays for the reward
    achieved_means = np.zeros(K_total)
    successfulRate_epoch = np.zeros(K_total)
    failDelay_epoch = np.zeros(K_total)
    
    size_sub6 = K_sub6 * N_sub6 * schedule_frameTime
    size_mmW = K_mmW * N_mmW * schedule_frameTime
    
    
    ##Start Opti

    for i in range(max_epoch):
        #print("iter : ", i)
        
        schedule_init_sub6 = np.zeros(size_sub6)
        schedule_init_mmW = np.zeros(size_mmW)
        
        schedule_0_sub6 = np.zeros(size_sub6)
        schedule_0_mmW = np.zeros(size_mmW)
     

        SINR_sub6_epoch = np.zeros(N_sub6)
        if(K_sub6 != 0):
            SINR_sub6_epoch = SINR_sub6[:, i]
    
        SINR_mmW_epoch = np.zeros(N_mmW)
        if(K_mmW != 0):
            SINR_mmW_epoch = SINR_mmW[:, i]
            
        j = 0
    
        ### mmWave & Sub-6 constraints
        cons_sub6 = []
        cons_mmW = []
        cons_sub6 = AllCons(schedule_init_sub6, SINR_sub6_epoch, Q_inv_sub6, l, K_sub6, N_sub6, users_infos_sub6)
        cons_mmW = AllCons(schedule_init_mmW, SINR_mmW_epoch, Q_inv_mmW, l, K_mmW, N_mmW, users_infos_mmW)
        
        ## Iterative optimization
        while j < J_max:          
            ## SUB 6
            if K_sub6 != 0:
                sol = minimize(U_bar, schedule_0_sub6, args=(schedule_init_sub6, SINR_sub6_epoch, Q_inv_sub6, l, K_sub6), method = 'COBYLA', constraints = cons_sub6, options = {'maxiter':30})

                scheduleOpt_sub6 = sol.x
                
                for k in range(K_sub6 * N_sub6 * schedule_frameTime):
                    schedule_init_sub6[k] = scheduleOpt_sub6[k]
    
                cons_sub6 = AllCons(schedule_init_sub6, SINR_sub6_epoch, Q_inv_sub6, l, K_sub6, N_sub6, users_infos_sub6)
            
            ## mmW
            if K_mmW != 0:
                sol = minimize(U_bar, schedule_0_mmW, args=(schedule_init_mmW, SINR_mmW_epoch, Q_inv_mmW, l, K_mmW), method = 'COBYLA', constraints = cons_mmW, options = {'maxiter':30})
            
                scheduleOpt_mmW = sol.x
                
                for k in range(K_mmW * N_mmW * schedule_frameTime):
                    schedule_init_mmW[k] = scheduleOpt_mmW[k]
    
                cons_mmW = AllCons(schedule_init_mmW, SINR_mmW_epoch, Q_inv_mmW, l, K_mmW, N_mmW, users_infos_mmW)
            j += 1
    
        result_sub6 = schedule_init_sub6
        result_mmW = schedule_init_mmW

        result_sub6[result_sub6 < 0.5] = 0
        result_sub6[result_sub6 > 0.5] = 1
        
        result_mmW[result_mmW < 0.5] = 0
        result_mmW[result_mmW > 0.5] = 1

        current_achieved = np.zeros(K_total)
        #Compute the achieved rate :
        if(K_sub6 != 0):
            cap_sub6 = F_k(result_sub6, SINR_sub6_epoch, K_sub6)
            disp_sub6 = V_k(result_sub6, SINR_sub6_epoch, Q_inv_sub6, l, K_sub6)
            
        if(K_mmW != 0):
            cap_mmW = F_k(result_mmW, SINR_mmW_epoch, K_mmW)
            disp_mmW = V_k(result_mmW, SINR_mmW_epoch, Q_inv_mmW, l, K_mmW)

        cpt_s6 = 0
        cpt_mmW = 0
        for k in range(K_total):
            if sub6_users[k] == 1:
                current_achieved[k] = cap_sub6[cpt_s6] - disp_sub6[cpt_s6]
                cpt_s6 += 1
            if mmW_users[k] == 1:
                current_achieved[k] = cap_mmW[cpt_mmW] - disp_mmW[cpt_mmW]
                cpt_mmW += 1
        
        #Update the mean achieved rate for this epoch's solution
        achieved_means = achieved_means + ((current_achieved - achieved_means) / (i+1))
        successfulRate_epoch = successfulRate_epoch + (achieved_means < users_infos[1,:])
        failDelay_epoch = failDelay_epoch + (current_achieved < users_infos[1,:])
    
        #Update the user requirements
        cpt_sub6 = 0
        cpt_mmW = 0
        for x in range(K_total) : 
            if(sub6_users[x] == 1):
                users_infos_sub6[cpt_sub6,1] = (i+2) * max(0, users_infos[1,x] - achieved_means[x]) + users_infos[1,x]
                cpt_sub6 += 1
            if(mmW_users[x] == 1):
                users_infos_mmW[cpt_mmW,1] = (i+2) * max(0, users_infos[1,x] - achieved_means[x]) + users_infos[1,x]
                cpt_mmW += 1

    achieved = achieved_means
    prob_rates = successfulRate_epoch / max_epoch
    prob_delay = failDelay_epoch / max_epoch
    time_opti = time.time() - start_time
    
    #print("achieved dans opti: ", achieved)
    #print("prob rates : ", prob_rates)
    #print("prob delay : ", prob_delay)
    
    return achieved, prob_rates, prob_delay, time_opti

#a = np.array([2, 2, 2, 2])
#scheduler(a)
```

# Replay Buffer to store all previously played episodes

```python
class ReplayBuffer(object):
    
    def __init__(self, max_size, input_shape, n_actions, discrete = True):
        self.mem_size = max_size
        self.input_shape = input_shape
        self.discrete = discrete
        self.mem_cntr = 0
        self.state_memory = np.zeros((self.mem_size, input_shape))
        self.new_state_memory = np.zeros((self.mem_size, input_shape))
        
        dtype = np.int32 if self.discrete else np.float32
        self.action_memory = np.zeros((self.mem_size, (3**K) - 1), dtype = dtype)
        self.reward_memory = np.zeros(self.mem_size)
        self.terminal_memory = np.zeros(self.mem_size, dtype = np.float32)
        
    def store_transition(self, state, action, reward, next_state, done):
        index = self.mem_cntr % self.mem_size
        self.state_memory[index] = state
        self.new_state_memory[index] = next_state
        self.reward_memory[index] = reward
        self.terminal_memory[index] = 1 - int(done)
        if self.discrete:
            #Discretize // Find discrete representation of actions
            actions = np.zeros((3**K) - 1)
            actions[action] = 1.0
            self.action_memory[index] = actions
        else:
            self.action_memory[index] = action
        self.mem_cntr +=1
    
    def sample_buffer(self, batch_size):
        max_mem = min(self.mem_cntr, self.mem_size)
        batch = np.random.choice(max_mem, batch_size)
        
        states = self.state_memory[batch]
        next_states = self.new_state_memory[batch]
        rewards = self.reward_memory[batch]
        actions = self.action_memory[batch]
        terminal = self.terminal_memory[batch]
        
        return states, actions, rewards, next_states, terminal

def build_dqn(lr, n_actions, input_dims, fc1_dims, fc2_dims, fc3_dims):
    model = Sequential([
                Dense(fc1_dims, input_shape=(input_dims, )),
                Activation('relu'),
                Dense(fc2_dims),
                Activation('relu'),
                Dense(fc3_dims),
                Activation('relu'),
                Dense(n_actions)])
    model.compile(optimizer = Adam(lr = lr), loss ='mse')
    
    return model

def to_ternary(val):
    arr = []
    while val:
        val, r = divmod(val,3)
        arr.append(r)
    arr = np.flip(arr)
    to_ret = np.append(np.zeros(K - len(arr)), arr).astype(np.int)  #Au cas où on a pas choisi assez d'actions, on complète avec des 0
    
    return to_ret.tolist()
```

#  Learning Agent

```python
class Agent(object):
    def __init__(self, alpha, gamma, n_actions, epsilon, batch_size,
                    input_dims, epsilon_dec=0.99, epsilon_end=0.01,
                    mem_size = max_mem_size, fname_input='models/dqn_model_K4.h5', fname_output='models/dqn_model_K4_pretrained.h5'):
        self.action_space = [i for i in range(n_actions)] # Might Have to change with action representation        self.gamma = gamma
        self.gamma = gamma
        self.epsilon = epsilon
        self.epsilon_dec = epsilon_dec
        self.epsilon_min = epsilon_end
        self.batch_size = batch_size
        self.model_file_input = fname_input
        self.model_file_output = fname_output
        
        self.memory = ReplayBuffer(mem_size, input_dims, n_actions,
                                    discrete = True)
        self.q_eval = build_dqn(alpha, n_actions, input_dims, 64, 128, 128)
        
    def remember(self, state, action, reward, next_state, done):
        self.memory.store_transition(state, action, reward, next_state, done)
        
    def choose_action(self, state):
        state = state[np.newaxis, :] #Rajoute une dimension à user_infos
        rand = np.random.random()
        if rand < self.epsilon:
            print('random action with epsilon: ', self.epsilon)
            action = np.random.choice(self.action_space)
        else:
            print('predicted action with epsilon: ', self.epsilon)
            actions = self.q_eval.predict(state)
            action = np.argmax(actions[0]) # Maybe actions[0] 
        return action
    
    def learn(self):
        if self.memory.mem_cntr < self.batch_size:
            return
        #if self.memory.mem_cntr > 3*self.batch_size:
        #    self.batch_size = 2*self.batch_size
        
        state, action, reward, next_state, done = self.memory.sample_buffer(self.batch_size)
        
        action_values = np.array(self.action_space, dtype = np.int8)
        action_indices = np.dot(action, action_values)
        
        q_eval = self.q_eval.predict(state)
        q_next = self.q_eval.predict(next_state)
        
        q_target = q_eval.copy()
        
        batch_index = np.arange(self.batch_size, dtype = np.int32)
        
        print(q_target.shape)
        q_target[batch_index, action_indices] = reward + self.gamma*np.amax(q_next, axis = 1)*done
        
        
        _ = self.q_eval.fit(state, q_target, verbose = 0)
        
        
        self.epsilon = self.epsilon*self.epsilon_dec if self.epsilon > self.epsilon_min else self.epsilon_min
        
    def save_model(self):
        self.q_eval.save(self.model_file_output)
    
    def load_model(self):
        self.q_eval = load_model(self.model_file_input)
```

```python
class Environment(object):
    def __init__(self, users_infos_filename, los_filename, target_rate_relab, target_delay_relab):
        self.actions_taken = 0
        self.infos_filename = users_infos_filename
        self.los_filename = los_filename
        self.users_infos = np.ravel(scipy.io.loadmat(self.infos_filename)['users_infos'][:7], order = 'F')
        self.users_infos[0::7] = np.ravel(scipy.io.loadmat(self.los_filename)['los_central_users'], order = 'F')
        print(self.users_infos)
        self.total_users = int(np.size(self.users_infos)/4)
        
        self.target_rate_relab = target_rate_relab
        self.target_delay_relab = target_delay_relab
        self.rate_outage_history = []
        self.delay_outage_history = []
        
    def reset(self):
        self.actions_taken = 0
        self.users_infos = np.ravel(scipy.io.loadmat(self.infos_filename)['users_infos'][:7], order = 'F')
        self.users_infos[0::7] = np.ravel(scipy.io.loadmat(self.los_filename)['los_central_users'], order = 'F')
        print(self.users_infos)
        self.total_users = int(np.size(self.users_infos)/4)  #Can Maybe Comment this line
        
        self.rate_outage_history = []
        self.delay_outage_history = []
        return self.users_infos
    
    
        
    # Run Matlab code for given action as argument    
    def step(self, action):
        self.actions_taken = self.actions_taken + 1
        print(self.actions_taken)     
        ternary_action = to_ternary(action)
        print(ternary_action)
        
        
        print("Début Optim")
        
        
        achieved, rate_probas, delay_probas, time_opti  = scheduler(ternary_action)
        
        print('achieved: \n', achieved)
        print('rate_probas: \n', rate_probas)
        print('delay_probas: \n', delay_probas)
        print('time_opti: \n', time_opti)
        
        #Transformation en 1D
        ach = np.asarray(achieved)
        relab_rate = 1 - np.asarray(rate_probas)
        relab_delay = 1 - np.asarray(delay_probas)
        
        self.rate_outage_history.append(np.asarray(rate_probas))
        self.delay_outage_history.append(np.asarray(delay_probas))
        
        self.users_infos[2::7] = ach 
        
        reward = - np.sum( self.users_infos[3::7] * (1 - relab_rate) + self.users_infos[4::7] * (1 - relab_delay))
        self.users_infos[3::7] = np.maximum(0, self.users_infos[3::7] + self.target_rate_relab - relab_rate)
        self.users_infos[4::7] = np.maximum(0, self.users_infos[4::7] + self.target_delay_relab - relab_delay)
        #self.weights_reward_rate = np.maximum(0, 1 + 1 - relab_rate)
        #self.weights_reward_delay = np.maximum(0, 1 + 1 - relab_delay)
        
        print('users_infos updated: \n', self.users_infos)
        print('updated weights:')
        print(self.users_infos[3::7])
        print(self.users_infos[4::7])
        
        success_rate = np.all(self.users_infos[2::7] >= self.users_infos[1::7])
        success_relab = np.all(np.logical_and((relab_rate >= 0.3) , (relab_delay >= 0.3)))
        success = success_rate and success_relab
        done = 1 if self.actions_taken >= 20 else success
        reward += success*1000
        print('Reward: \n', reward)
        
        return self.users_infos, reward, done, time_opti, None
```

```python
agent = Agent(gamma = 0.2, epsilon =1.0, epsilon_dec=0.95, alpha = 0.1, input_dims = 7*K,
                  n_actions = (3**K)-1, mem_size = max_mem_size, batch_size=64, epsilon_end = 0.1)

users_infos_filename = 'MATLAB/ressources/users_infos_K4N4.mat'
los_filename = 'MATLAB/ressources/topologies/topo_3x3/4users/custom.mat'
env = Environment(users_infos_filename, los_filename, target_rate_outage, target_delay_outage)

total_reward_history = []
avg_reward_history = []
avg_sum_rate = []
max_sum_rate = []
eps_history = []
actions_until_done = []
rate_outage = []
delay_outage = []

time_history_DQN = []
time_history_Opti = []
total_time = 0
total_time_DQN = 0
total_time_Opti = 0


for i in range(n_games):
    print('\n This is a New Episode STARTING: \n')
    start_time = time.time()
    time_opti_loop = 0
    done = False
    total_reward = 0
    sum_rate = []
    observation = env.reset() #Observation est un vecteur contenant user_infos
    while not done:
        
        action = agent.choose_action(observation)
        next_observation, reward, done, time_opti, info = env.step(action) # TODO CREATE THIS 
        time_opti_loop += time_opti
        total_reward += reward
        sum_rate.append(sum(next_observation[2::7]))
        agent.remember(observation, action, reward, next_observation, done)
        observation = next_observation
        agent.learn()
    
    eps_history.append(agent.epsilon)
    total_reward_history.append(total_reward)
    avg_sum_rate.append(np.mean(sum_rate))
    max_sum_rate.append(np.max(sum_rate))
    actions_until_done.append(env.actions_taken)
    rate_outage.append(np.mean(env.rate_outage_history, axis = 0))
    delay_outage.append(np.mean(env.delay_outage_history, axis = 0))

        
    avg_reward = np.mean(total_reward_history[max(0, i - 20):(i+1)])
    avg_reward_history.append(avg_reward)
    
    
    period_time_DQN = time.time() - start_time - time_opti_loop
    
    time_history_DQN.append(period_time_DQN)
    time_history_Opti.append(time_opti_loop)
    
    total_time = total_time + period_time_DQN + time_opti_loop
    total_time_DQN = total_time_DQN + period_time_DQN
    total_time_Opti = total_time_Opti + time_opti_loop
    
    print('episode ', i, 'score %.2f' % total_reward, 'average score %.2f' % avg_reward)
    
    #if i%10 == 0 and i > 0:
        #agent.save_model()   
        
print("\n FINI \n")
```
